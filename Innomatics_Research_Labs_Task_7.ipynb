```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False

    def set_state(self, action):
        if action.upper() == "ON":
            self.is_on = True
        elif action.upper() == "OFF":
            self.is_on = False

        self.display_status()

    def display_status(self):
        status = "ON" if self.is_on else "OFF"
        print(f"{self.name} is {status}")

# Input
light_name = "Bedroom Light"
action = "ON"

my_light = SmartLight(light_name)
my_light.set_state(action)
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

emp1 = Employee("Rahul", "EMP102", "IT")

# Displaying ID card
emp1.display_id_card()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

# Input
name_input = "Anita"
phone_input = "9876543210"

new_contact = Contact(name_input, phone_input)

# Displaying output
new_contact.display_contact()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

# Input
product_name = "Headphones"
product_price = 2499

item = Product(product_name, product_price)

# Displaying price tag
item.print_price_tag()
```

```python
class Movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def display_details(self):
        print(f"Movie: {self.title}")
        print(f"Rating: {self.rating} / 5")

# Input
movie_name = "Inception"
movie_rating = 4.8

film = Movie(movie_name, movie_rating)

# Displaying output
film.display_details()
```

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

# Input
name = "Suman"
location = "Hyderabad"
order_label = Delivery(name, location)

# Displaying output
order_label.print_delivery_details()
```
